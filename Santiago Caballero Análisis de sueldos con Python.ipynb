1. Descargar, leer y unir en una sola estructura de datos. ¿Qué estructura elegiste y por qué?

````python
print( '''
En primer lugar, utilice listas para poder abrir los archivos y volcarlos los datos que estos contienen.
Elegi esta opción ya que Las listas son mutables y esto permite realizar multiples operaciones para poder
ajustar la base que es extraida de los archivos, previo a su análisis. 
Una vez realizado estos primeros ajustse, ya para presentar la estructura de datos completa con la 
referencia a cada una de las categorías, conforme una lista de diccionarios donde cada diccionario
corresponde a una observación/individuo, en el cual las keys representan la pregunta y los valores
las respuestas aplicado a cada observación, lo que es visualmente mas claro.
En el mismo diccionario, la strings de las keys son inmutables, lo que es valido para el análisis
porque se corresponde con cada pregunta y estas no se modifican, y por otro lado los valores de 
las respuestas si son mutables, lo que sirve para realizar ajustes en caso de que la observacion
haya sido mal registrada, o que el contenido deba ser modificado por alguna otra razón.
''')
````

````python
#descarga de archivos con datos brutos originales de la encuesta 
archivo_ARG = '2021.2 - sysarmy - Encuesta de remuneración salarial Argentina - Argentina.csv'
archivo_LATAM = '2021.2 - sysarmy - Encuesta de remuneración salarial Latam - Latam.csv'
import csv

# lectura de archivos e inclusion en listas predefinidas
datos_ARG = []

with open(f"{archivo_ARG}","r",encoding='utf8') as f1:
    reader = csv.reader(f1)
    for linea in reader:
        datos_ARG.append(linea)

del datos_ARG[:10]

for i in range(len(datos_ARG)):
    datos_ARG[i].append('Argentina')

for i in range (0,len(datos_ARG)):
    datos_ARG[i].append(datos_ARG[i][0])
    datos_ARG[i].remove(datos_ARG[i][0])
    
datos_LATAM = []

with open(f"{archivo_LATAM}","r",encoding='utf8') as f2:
    reader = csv.reader(f2)
    for linea in reader:
        datos_LATAM.append(linea)

del datos_LATAM[:11]

for i in range (len(datos_LATAM)):
    datos_LATAM[i].append(datos_LATAM[i][0])
    datos_LATAM[i].remove(datos_LATAM[i][0])
    
region_LATAM = []
for obs in datos_LATAM:
    for i in range(0,18):
        if len(obs[i]) != 0:
            region_LATAM.append(obs[i])

for i in range (len(datos_LATAM)):
    datos_LATAM[i].append(region_LATAM[i])
    del(datos_LATAM[i][0:18])

#Uno la lista de Argentina con la de LATAM conformando una única lista con todas las respuestas
datos = datos_ARG + datos_LATAM

'''
creo una lista de diccionarios, donde las keys representan cada pregunta de la encuesta 
y los valores cada respuesta
'''

base = []
for i in datos:
    base.append({
        "Tipo de contrato": "",
        "Salario Bruto": "",
        "Salario Neto" : "",
        "Pagos en dólares": "",
        "Tipo de cambio": "",
        "Conformidad con salario": "",
        "Sueldo respecto a semestre": "",
        "Bono": "",
        "Razon del bono":"",
        "Ajustes por inflacion":"",
        "% del ajuste":"",
        "Ultimo ajuste":"",
        "Ocupacion": "",
        "Años de experiencia": "",
        "Años en empresa actual": "",
        "Años en puesto actual": "",
        "Personas a cargo":"",
        "Plataformas":"",
        "Lenguajes de programación": "",
        "Framework, herramientas y librerias":"",
        "Bases de datos":"",
        "QA / Testing": "",
        "IDEs":"",
        "Cantidad de personas en organización":"",
        "Trabaja para empresa sin oficina en ciudad":"",
        "Actividad principal":"",
        "Recomendacion como buen lugar para trabajar": "",
        "Calificacion de politicas de diversidad e inclusion": "",
        "Beneficios extra":"",
        "Salir o seguir contestando 1":"",
        "Nivel de estudios alcanzado": "",
        "Estado del nivel": "",
        "Carrera": "",
        "Universidad":"",
        "Cursos de especialización": "",
        "Contribucion a proyectos Open Source": "",
        "Programación como hobbie": "",
        "Salir o seguir contestando 2":"",
        "Tiene guardias":"",
        "Ingreso por guardias":"",
        "% bruto o neto":"",
        "Salir o seguir contestando 3":"",
        "Situación de violencia laboral":"",
        "Persona con discapacidad": "",
        "Dificultades en conseguir trabajo por discapacidad":"",
        "Salir o seguir contestando 4":"",
        "Vinculo con pandemia":"",
        "Hijos menores de edad":"",
        "Cantidad de personas conviviendo":"",
        "Personas conviviendo":"",
        "Espacio para trabajar":"",
        "Comparte el equipo de trabajo":"",
        "Tipo de cuarentena":"",
        "Cambio situacion laboral con pandemia":"",
        "Apoyo de empresa en pandemia":"",
        "Afectación de empresas por pandemia":"",
        "Beneficio nuevo por pandemia":"",
        "Edad": "",
        "Identidad de Género": "", "Pais":"", "Region":""
       })

for i, j in enumerate (datos):
    for k, l in enumerate(base[i].keys()):
        base[i][l] = datos[i][k]
````

2. Printear porcentaje de participación por región, ordenado de mayor a menor. Ej: 

```
- Ciudad Autónoma de Buenos Aires - 59.4%
- Formosa - 0.04%
```

````python
#extraigo los regiones de la lista de diccionarios y creo una lista con las regiones
region = []
for i in range (len(base)):
    region.append(base[i]['Region'])

#creo una lista y defino una funcion que cuente observaciones de cada region y agregue esos numeros a la lista    
region_count = []
def contar_region (a):
    region_count.append(region.count(a))

#creo una variable con los nombres de las regiones y las agrego a la lista anterior
region_keys = list(set(region))    
for i in range (len(region_keys)):
    contar_region(region_keys[i])

#transformo los valores en porcentajes
region_count = [x/len(region)*100 for x in region_count]
region_count = [round(x,2) for x in region_count]

# hago un diccionario combinando la lista de nombres y el porcenaje de participacion de cada region
dict_region = dict(zip(region_keys,region_count))

#utilizo un operador que permite ordenar el diccionario y presentar los resultados
import operator
dict_region_sort = sorted(dict_region.items(), key=operator.itemgetter(1), reverse=True)

print('PORCENTAJE DE PARTICIPACION POR REGION:')
for key, value in dict_region_sort:
    print (key, '-', value, '%')  
````

8. Formatear las carreras universitarias:

- Reemplazar `ñ` por `n`
- *lic*, *lic.* por Licenciatura
- *tec*, *tec.* por Tecnucatura
* *cs, *cs.* por Ciencias
* *ed, ed.* por Educación
* Transformaciones que se consideren necesarias

Printear porcentaje según carrera
